### Изменение типов данных

*Изменим тип данных в столбце с годом релиза с вещественного на целочисленный, чтобы не было искажений как 2006,3 год.*

```python
# Комментарий ревьюера
temp = data.copy() # Создаем копию датасета
```

```python
data['year_of_release'] = data['year_of_release'] .astype('Int64')
```

### Обработка дубликатов

<u>Сохраним исходную размерность датафрейма</u>

```python
rows_before = data.shape[0]
```

<u>Обработка явных дубликатов</u>

*Перед удалением явных дубликатов переведем названия в нижний регистр, чтобы устранить соответсвующие возможные дубликаты.*

```python
data['name'] = data['name'].str.lower()
```

```python
data = data.drop_duplicates()
```

<u>Обработка неявных дубликатов</u>

*Рассмотрим возможные ошибки (неяные повторения, как например Action и action, Sports и спорт) в названиях плтформы и жанрах.*

```python
columns = ['platform', 'genre', 'rating']

for column in columns:
  print('\n')
  print(column, data[column].unique())
```

*Неявных дубликатов не обнаружено, ошибок в форме записи также не найдено.*

```python
data['genre'] = data['genre'].str.lower() #переведем жанры в нижний регистр 
```

*Удалим неявные дубликаты среди названий, платформ и даты релиза.*

```python
data = data.drop_duplicates(subset=['name', 'platform', 'year_of_release'], keep=False)
```

`year_of_release`: *посмотрим возможные ошибки в форме записи года и заодно ознакомимся с временным периодом архивов магазина, а также самым популярным годом выхода игр.*

```python
data.groupby('year_of_release')['year_of_release'].count()
```

*Таким образом, база данных за 1980-2016 года. Больше всего вышло игр за 2008 год.*

### Обработка пропусков

<u>Подчет пропусков</u>

```python
data.isna().sum()
```

<u>Столбцы name и genre</u>

*В обоих столбцах (природа пропуска скорее всего одинаковая) пропущено всего две строки, значения никак не подлежат восстановлению, поэтому просто избавимся от этих строк, что никак не исказит дальнейший анализ. Вероятно пропуски связаны с технической ошибкой, возможно игры настолько непопулярны, что название и жанр были утеряны, а возможно утерены данные при их переносе из архива.*

```python
data = data.dropna(subset=['name', 'genre'])
```

<u>Столбцы critic_score и users_score</u>

```python
data['critic_score'].unique()
```

*В столбце `critic_score` не будем проводить заполнение пропусков, так как это исказит данные. Рейтинг может отсутствовать из-за допустим года выпуска игры (слишком старая игра), недостаточного количества оценок, чтобы составить рейтинг (из-за непопулярности игры), а также из-за страны производителя. Аномальных значений более 100 не обнаружено, как и отрицательных.*

```python
data['user_score'].unique()
```

```python
data[['user_score', 'critic_score']] = data[['user_score', 'critic_score']].fillna(-1)
```

*Аналогично для столбца `user_score`. Также игроки могли не оценить игру из-за ее непопулярности, относительно недавнего года выхода. Здесь также встречается значение 'tbd' -  to be determined/decided (будет определено/решено), то есть такое значение просто отсутвует в базе. Заменим на `-1`, чтобы не искажать анализ. Аномальных значений выше 10 не обнаружено, как и отрицательных.*

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = -1
```

```python
data['user_score'] = data['user_score'].astype('float') #поменяем тип данных со строчного на вещественный
```

<u>Столбец rating</u>

*Рейтинг в базе данных составляется организацией ESRB. Согласно [Wikipedia](https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board), Entertainment Software Rating Board — негосударственная организация, основное направление деятельности — принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в **США** и **Канаде**. [DTF](https://dtf.ru/gamedev/265535-vozrastnoj-rejting-i-kak-ego-poluchit): "Есть несколько основных возрастных рейтингов: американская ESRB, европейская PEGI, немецкая USK, японская CERO, российская RARS, а также еще несколько региональных организаций". Скорее всего отсутствующие данные в строках связаны со страной производителя, не все в первую очередь оцениваются по американской и канадской системам. Возможна также техническая ошибка (за поздние года не сохранились рейтинги/их не обновили, ошибка в переносе данных).*

*Заменим на `no_ESRB`, так как вероятно пропуск связан с отсутвием рейтинга для США и Канады.*

```python
data['rating'] = data['rating'].fillna('no_ESRB')
```

<u>Столбец year_of_release</u>

*Таких пропусков всего 229 среди более 16000 строк, поэтому просто отбросим строки с пропусками.*

*Конечно, возможно частично заполнить, взяв год релиза из столбца `name`, но такой маневр не рацианален, так как пропусков около 2%.*

```python
data = data.dropna(subset=['year_of_release'])
```

*Посчитаем сумарные продажи, создадим отдельный столбец.*

```python
warnings.filterwarnings("ignore")

data['global_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + \
data['other_sales']
```

### Год выпуска игр

*Cколько игр выпускалось в разные годы. Важны ли данные за все периоды?*

*Для описания дискретных величин лучше подходит столбчатая диаграмма.*

```python
data.groupby('year_of_release')['year_of_release'].count().plot.bar(figsize=(12,6), grid=True, 
                                                                    color='green', alpha=0.4
)
plt.title('Годы выпуска игр', fontsize=16)
plt.xlabel('Год выпуска', fontsize=12)
plt.ylabel('Количество продаж', fontsize=12)
plt.show()
```

*Как было замечено ранее, больше всего игр за 2008. Как видно,игр за разные года девяностых не более 400. Из графика видно, что с начала нулевых игр значительно больше выпускалось, но, начиная с 2009 года, наблюдается снижение объема выпуска компьютерных игр.*

*Прочитав информацию из [Wikipedia](https://ru.wikipedia.org/wiki/%D0%98%D1%81%D1%82%D0%BE%D1%80%D0%B8%D1%8F_%D0%BA%D0%BE%D0%BC%D0%BF%D1%8C%D1%8E%D1%82%D0%B5%D1%80%D0%BD%D1%8B%D1%85_%D0%B8%D0%B3%D1%80), можно прийти к следующему заключению*:
- *в 1980-х только зараждался интерес к компьютерным играм, появлялись аркадные автоматы, игровые консоли и домашние компьютеры*. Многие игры, как например легендарная Ms. Pac-Man, перешли из аркадрных автоматов на компьютеры;
- *в 1990-х произошел полный переход на 3D-графику, появились новые игровые жанры: шутер от первого лица, стратегия в реальном времени и массовая многопользовательская онлайн-игра (ММО). Постепенно набирали популярность игровые приставки такие, как Nintendo, Playstation, XBOX.*;

*Для платформы PS3 наблюдается слабая корреляция между глобальными продажами и оценками от критиков - около 0,03, а корреляция с оценками от пользователя близка к 0. Основные оценки в диапазоне от 6-8. При этом корреаляция глобальных продаж с отзывами от критиков низкая - около 0,4. Основные оценки критиков в районе 68-84. Посмотрим, как обстоят дела у других топовых платформ. Не будем строить диаграмму рассеивания, сделаем вывод по коэффициенту корреляции.*.

```python
for platform in best_sellers:
    
    platforms_user = streamchik.loc[(streamchik['platform'] == platform) & (streamchik['user_score']>=0)]
    platforms_critic = streamchik.loc[(streamchik['platform'] == platform) & (streamchik['critic_score']>=0)]

    user_corr = platforms_user['user_score'].corr(platforms_user['global_sales'])
    critic_corr = platforms_critic['critic_score'].corr(platforms_critic['global_sales'])
    
    # подсчет и оценка корреляции между отзывами и глобальными продажами
    
    if 0.1 > user_corr >= 0:
        print('Корреляция между отзывами пользователей и игровой платформой ', platform, \
              'отсутствует -', user_corr.round(3))
    if 0.5 > user_corr >= 0.1:
        print('Корреляция между отзывами пользователей и игровой платформой ', platform, \
              ': Низкая -', user_corr.round(3))
    if 0.8 >= user_corr >= 0.5:
        print('Корреляция между отзывами пользователей и игровой платформой ', platform, \
              ': Средняя -', user_corr.round(3))
    if 1 >= user_corr > 0.8:
        print('Корреляция между отзывами пользователей и игровой платформой ', platform, \
              ': Высокая -', user_corr.round(3))
    print('\n')
    
    if -0.1 < user_corr < 0:
        print('Корреляция между отзывами пользователей и игровой платформой ', platform, \
              'отсутствует -', user_corr.round(3))
    if -0.5 < user_corr <= -0.1:
        print('Обратная корреляция между отзывами пользователей и игровой платформой ', platform, \
              ': Низкая -', user_corr.round(3))
    if -0.8 <= user_corr <= -0.5:
        print('Обратная корреляция между отзывами пользователей и игровой платформой ', platform, \
              ': Средняя -', user_corr.round(3))
    if -1 <= user_corr < -0.8:
        print('Обратная корреляция между отзывами пользователей и игровой платформой ', platform, \
              ': Высокая -', user_corr.round(3))
    print('\n')
    
    if 0>= critic_corr > 0.1:
        print('Корреляция между отзывами критиков и игровой платформой ', platform, \
              'отсутствует -', critic_corr.round(3))
    if 0.5 > critic_corr >= 0.1:
        print('Корреляция между отзывами критиков и игровой платформой ', platform, \
              ': Низкая -', critic_corr.round(3))
    if 0.8 >= critic_corr >= 0.5:
        print('Корреляция между отзывами критиков и игровой платформой ', platform, \
              ': Средняя -', critic_corr.round(3))
    if 1 >= critic_corr > 0.8:
        print('Корреляция между отзывами критиков и игровой платформой ', platform, \
              ': Высокая -', critic_corr.round(3))
    print('\n') 
    
    
    # построение диаграммы рассеяния в одних координатах
    fig, ax = plt.subplots(1 ,2, figsize=(12,6))
    sns.scatterplot(data=platforms_user, x='user_score', y='global_sales', ax=ax[0])
    sns.scatterplot(data=platforms_critic, x='critic_score', y='global_sales', ax=ax[1])
    fig.suptitle(platform, fontsize=16)
    ax[0].set(xlabel='Рейтинг пользователей')
    ax[1].set(xlabel='Рейтинг критиков')
    ax[0].set(ylabel='Количество продаж')
    ax[1].set(ylabel='Количество продаж')
    plt.tight_layout()
    sns.set_style("whitegrid")
    plt.show()
    
```

*У PS4, Xbox One, PS3, X360, 3DS наблюдается низкая (или вовсе отсутвует) корреляция между глобальными продажами и отзывами как критиков, так и пользователей. Нельзя сказать, что пользователи покупают игры, ориентируясь только на отзывы. Нужно отметить, что соотношение глобальных продаж с рейтингом критиков несколько выше, чем с рейтингом пользователей.*

## Портрет пользователей каждого региона
<a id='portret'></a>

*Определим для пользователя каждого региона (NA, EU, JP) самые популярные плтформы и жанры, а также влияние рейтинга ESRB на продажи.*

*Напишем функцию для отрисовки круговых диаграмм.*

```python
def pie_users(parametr):
    
    countries = streamchik.columns[4:7] 
    fig, ax = plt.subplots(1, 3, figsize=(16, 6))
    colors = sns.color_palette('pastel')[0:6]
    title_1 = {'na_sales': 'Северная Америка', 'eu_sales': 'Европа', 'jp_sales': 'Япония'}

    
    for i in range(3):
        
        region_games = (streamchik.groupby(parametr)[countries[i]].sum().
                        reset_index().
                        sort_values(by=countries[i], ascending = False, ignore_index = True).
                        head())
        display(region_games.style.hide_index())
        if streamchik[countries[i]].sum() - region_games[countries[i]].sum() >0:
            region_games = region_games.append({parametr: 'Other', countries[i]: 
                                                streamchik[countries[i]].sum() - region_games[countries[i]].sum()}, 
                                                ignore_index= True)
        ax[i].pie(region_games[countries[i]], labels = region_games[parametr], colors=colors, autopct='%1.1f%%')
        ax[i].set_title(title_1[region_games.columns[1]]) 
    
    fig.suptitle('Портрет пользователя каждого региона', fontsize=16)
    plt.show()
```

## Проверка гипотез
<a id='hypotheses'></a>

### Средние пользовательские рейтинги платформ Xbox и PC
<a id='hypothesis_platform'></a>

*Начальная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC различны.*

*Была сформулирована двусторонняя альтернативная теория, так как сразу нельзя сказать у какой платформы выше/ниже рейтинги. Для проверки проведем Т-тест для сравнения двух генеральныъх совокупностей методом `scipy.stats.ttest_ind()`.*

*Для начала сравним дисперсии выборок, чтобы сравнить их однородность выборок. В нашем случае, когда размерность выборки достаточно большая (более 30 значений), возможно допущение и без проверки о том, что дисперсии считаются равными, тест объединит обе выборки в одну, чтобы эту дисперсию поточнее оценить.*

```python
sample_1 = streamchik[(streamchik['platform']=='XOne') & (streamchik['user_score']>=0)]['user_score']
sample_2 = streamchik[(streamchik['platform']=='PC') & (streamchik['user_score']>=0)]['user_score']
```

```python
np.var(sample_1) #дисперсия для платформы XOne
```

```python
np.var(sample_2) #дисперсия для платформы PC
```

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу') 
```

```python
sample_1.mean() #среднее значение для платформы XOne
```

```python
sample_2.mean() #среднее значение для платформы PC
```

*Как видно, медианные значения практически совпадают.*

### Средние пользовательские рейтинги жанров Action и Sports
<a id='hypothesis_genre'></a>

*Начальная гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые. Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные.*

*Была сформулирована двусторонняя альтернативная теория, так как сразу нельзя сказать у какого жанра выше/ниже рейтинги. Для проверки проведем Т-тест для сравнения двух генеральныъх совокупностей методом `scipy.stats.ttest_ind()`.*

*Для начала сравним дисперсии выборок, чтобы сравнить их однородность выборок. В нашем случае, когда размерность выборки достаточно большая (более 30 значений), возможно допущение и без проверки о том, что дисперсии считаются равными, тест объединит обе выборки в одну, чтобы эту дисперсию поточнее оценить.*

```python
sample_3 = streamchik[(streamchik['genre']=='action') & (streamchik['user_score']>=0)]['user_score']
sample_4 = streamchik[(streamchik['genre']=='sports') & (streamchik['user_score']>=0)]['user_score']
```

```python
np.var(sample_3) #дисперсия для жанра action
```

```python
np.var(sample_4) #дисперсия для жанра sports
```

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_3, sample_4, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу') 
```

```python
sample_3.mean().round() #среднее значение для жанра action
```

```python
sample_4.mean() #среднее значение для жанра sports
```

*На уровне статистической значимости 5% есть основания отвергнуть нулевую гипотезу о средних значений пользовательских рейтингов жанров action и sports в пользу альтернативной гипотезы.* 

**Вывод**: начальные гипотезы были сформулированы таким образом, чтобы понять соответвуют ли выборки ожидаемому значению, то есть когда средние пользовательские рейтинги (как в случае с платформами, так и с жанрами) одинаковые у двух генеральных совокупностей. В качестве альтернативной гипотезы сформулирована двухсторонняя гипотеза, так как проделанный анализ не дает однозначного представления о том у какой платформы (какого жанра) будет выше/ниже рейтинг. Для проверки  проведен Т-тест, подходящий для распределений как нормальных, так и отличных от нормального в случае объема выборки более 30, для сравнения двух независимых генеральных совокупностей методом `scipy.stats.ttest_ind()`. 

В результате не отвергаем нулевую гипотезу: средние пользовательские рейтинги платформ Xbox One и PC одинаковые; не отвергаем альтернаятивную гипотезу: средние пользовательские рейтинги жанров Action и Sports разные.
