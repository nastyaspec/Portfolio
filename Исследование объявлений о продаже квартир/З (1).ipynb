### Заполнение пропусков

*Заполняем стобцы с числом балконов. Под пропуском понимаем полное отсутствие балконов.*

```python
data['balcony'] = data['balcony'].fillna(0)
```

*Заполняем пропуски в столбцах с площадями. Как видно, жилая площадь сильно зависит от общей (корреляция близка к 1).*

```python
#Найдем отношение жилой площади к общей
ratio_liv = (data['living_area'] / data['total_area']).median()
#Заполним пропуски
data.loc[data['living_area'].isna(), 'living_area'] = (
    (data.loc[(data['living_area'].isna()) \
              & (data['total_area'].index == data['living_area'].isna().index), 'total_area']) \
                   * ratio_liv
)
```

*Корреляция между площадью кухни и общей площади квартиры средняя и составляет 0,6, что выше, чем относительно других параметров.*

```python
#Найдем отношение площади кухонь к общей
ratio_kit = (data['kitchen_area'] / data['total_area']).median()
#Заполняем пропуски
data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = (
    (data.loc[(data['kitchen_area'].isna()) \
              & (data['total_area'].index == data['kitchen_area'].isna().index), 'total_area']) \
                   * ratio_kit
)
```

*Заполняем пропуски в столбце с булевыми значениями `is_apartment`. Пропуски есть практически во всех строках: 20924 из 23699. Предполагаем, что пользователь заполнял бланк исключительно при ложном утвержении.*

```python
data.groupby('is_apartment')['is_apartment'].count() #проверка какие значения обычно вписывали пользователи
```

```python
data['is_apartment'] = data['is_apartment'].fillna(True)
```

*Обрабатываем пропуски в `locality_name`. Таких пропусков немного - всего 49, скорее всего в пропусках виноват человеский фактор (невнимательность, неаккуратность пользователя и пр.). Пропусков в `floors_total` также немного - 86, но достаточной корреляции нет ни с одним параметром. Отбросим строки с пропусками в данных столбцах, что никак не отразится на качестве анализа.*

```python
data = data.dropna(subset = ['locality_name', 'floors_total'])
```

```python
data.isna().sum()
```

Пропуск в столбце `is_apartment` пропущены значения скорее всего из-за необязательности для заполнения поля в анкете; пользователи заполняли только в случае другого вида жилья. Данные о замерах квартиры (площадь, высота потолка) скорее всего пропущены из-за отсутвия данных из выписки из ЕГРН, самостоятельных замеров или утаяны пользователем. #### Изменение типов данных

*Заменим вещественный тип данных на целочисленный в столбцах, значения которых входят в множество натуральных чисел. Это предотвратит возможные ошибка, как ,например, 2,5 балкона. В случае со стоимостью продажи удобнее цены рассматривать целочисленные.* 

```python
type_to_change = ['balcony','parks_nearest', 'ponds_nearest', 'days_exposition']
data[type_to_change] = (
    data[type_to_change]
    .fillna(-1)
    .astype(int)
    .replace(-1, np.nan)
    .astype('Int64')
)
```
